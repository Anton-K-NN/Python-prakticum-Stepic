```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
xpl=np.arange(0.01,500,0.01) #положительные х

xotr=np.arange(-0.01,-500,-0.01) #отрицательные х
```

```python
def f(x,alfa,beta):
    f=(x**beta + alfa**beta)/x**beta
    return f
    
```

```python
#Задача 1. Часть 1.
```

```python
alfa1=1
beta1=1

alfa2=2
beta2=1

alfa3=1
beta3=2
```

```python
f1pl=f(xpl,alfa1,beta1)
f1otr=f(xotr,alfa1,beta1)

f2pl=f(xpl,alfa2,beta2)
f2otr=f(xotr,alfa2,beta2)

f3pl=f(xpl,alfa3,beta3)
f3otr=f(xotr,alfa3,beta3)
```

```python
fig1=plt.figure(figsize=(13,8),dpi=120)
plt.ylim(-20,20)
plt.xlim(-5,5)
plt.plot(xpl,f1pl,'b',label='функция f(x) при $\\alpha=1, \\beta=1$')
plt.plot(xotr,f1otr,'b')
plt.plot(xpl,f2pl,'r', label='функция f(x) при $\\alpha=2, \\beta=1$')
plt.plot(xotr,f2otr,'r')
plt.plot(xpl,f3pl,'g', label='функция f(x) при $\\alpha=1, \\beta=2$')
plt.plot(xotr,f3otr,'g')
plt.xlabel('Абсцисса - x')
plt.ylabel('Ордината - f(x)')
plt.title('График функции $ f(x)=\\frac{x^\\alpha + \\alpha^\\beta}{x^\\beta} $ \n \
при различных значениях параметров $\\alpha \:и\: \\beta$')
plt.legend()
plt.show()
plt.savefig('fig1.svg')
```

```python
'''
Выводы:

При положительных х, при стремлении  х к +0 исследуемая функция при всех значениях параметров стремится к +бесконечности.
Вертикальной асимптотой графиков является прямая y=0.
Причем при значениях параметров альфа=1 и бета=2 стремится к бесконечности (возрастостает) быстрее чем при прочих значениях параметров.
При значении парметров альфа=1 и бета =1 график функции медленнее всего стря к +бескочности.

Горизонтальной асимптотой графиков является прямая х=1.
исследуемая функция при всех значениях параметров стремится к 1 сверху.
По прежнему при при значениях параметров альфа=1 и бета=2 стремится к единице (убывает) быстрее чем при прочих значениях параметров.
ОДнако теперь медленнее всехубывает функция при значениях параметров альфа =2 и бета=2.

'''
```
